```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("netflix_titles.csv")
```

1. Count plot

```python
sns.countplot(x="type", data=df, palette="viridis")
plt.title("Movies vs TV shows")
plt.show()
```

2. Histogram

```python
sns.histplot(df["release_year"], bins=20, color="red")
plt.title("Release year distribution")
plt.show()
```

3. Box plot

```python
sns.boxplot(df["release_year"])
plt.title("Release year spread")
plt.show()
```

4. Violin Plot

```python
sns.violinplot(df["release_year"])
plt.show()
```

5. Bar plot

```python
sns.barplot(x="type", y="release_year", data=df, palette="pastel")
plt.title("Average release year by title")
plt.ylabel("Average release year")
plt.show()
```

6. Scatter Plot

```python
content_age= 2025- df["release_year"]
sns.scatterplot(x=df["release_year"], y=content_age)
plt.show()
```

7. Line Plot

```python
years= df["release_year"].value_counts().sort_index()
sns.lineplot(x=years.index, y=years.values)
plt.show()
```

8. Heatmap

```python
# Converting type to numeric
df["type_numeric"] = df["type"].map({"Movie": 0, "TV Show": 1})

# Creating correlation matrix
corr = df[["release_year", "type_numeric"]].corr()


sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation between Release Year and Type")
plt.show()
```

9. Pairplot

```python
sns.pairplot(df[["release_year"]])
plt.show()
```

10. KDE Plot

```python
# checking the density distribution
sns.kdeplot(df["release_year"])
plt.show()
```

11. Stacked category comparison

```python
sns.countplot(x="release_year", hue="type", data=df)
plt.xticks(rotation=90, fontsize=5)
plt.show()
```

12. Top countries visualization

```python
top_countries= df["country"].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.show()
```
