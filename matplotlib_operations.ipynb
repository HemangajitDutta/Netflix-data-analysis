```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= pd.read_csv("netflix_titles.csv")
```

1. Basic Plot Setup

```python
plt.figure(figsize=(10,5))
```

2. Line Plot

```python
years= df["release_year"].value_counts().sort_index()
plt.plot(years.index, years.values)
plt.xlabel("Year")
plt.ylabel("Number of titles")
plt.title('netflix content growth over time')
plt.show()
```

3. Bar Plot (Category comparison)

```python
counts= df["type"].value_counts()
plt.bar(counts.index, counts.values, color= ["red", "yellow"])
plt.xlabel("Type")
plt.ylabel("No of titles")
plt.title("Movies vs TV Shows")
plt.show()
```

4. Horizontal Bar Plot

```python
top_countries= df["country"].value_counts().head(10)
plt.barh(top_countries.index, top_countries.values, color= ["magenta", "orange", "blue", "yellow", "red", "pink", "black", "aqua", "green", "purple"])
plt.title("Top 10 countries by content")
plt.show()
```

5. Histogram

Release year distribution

```python
plt.hist(df["release_year"], bins=20, color="green")
plt.xlabel("Release year")
plt.ylabel("Frequency")
plt.title("Distribution of release years")
plt.show()
```

6. Box Plot (Spread & outliers)

```python
plt.boxplot(df["release_year"])
plt.title('Release year spread')
plt.show()
```

7. Scatter Plot (Relationship)

```python
content_age= 2025- df["release_year"]
plt.scatter(df["release_year"], content_age, color="red")
plt.xlabel("Release year")
plt.ylabel("Content Age")
plt.title("Release year vs content age")
plt.show()
```

8. Pie Chart

```python
plt.pie(counts.values, labels=counts.index, autopct="%1.1f%%", colors=["aqua", "orange"])
plt.title("Movie vs TV Shows")
plt.show()
```

9. Multiple plots & Subplots

```python
years= df["release_year"].value_counts().sort_index()
counts= df["type"].value_counts()
plt.figure(figsize=(9,9))
plt.subplot(2, 2, 1)
plt.plot(years.index, years.values)
plt.subplot(2, 2, 2)
plt.bar(counts.index, counts.values)
plt.show()
```

10. Customization

```python
plt.plot(years.index, years.values, linestyle="--", marker="o", markerfacecolor="red", markeredgecolor="black")
plt.xticks(rotation=45)
plt.yticks(rotation=90)
plt.grid(True)
```

11. Plot from GroupBy Data

```python
year_type= df.groupby(["release_year", "type"]).size().unstack()
year_type.plot(kind="line")
plt.title("Movies vs tv shows over time")
plt.show()
```
