14. Combining Data

```python
df1= pd.DataFrame({
    "title": ["Movie A", "Movie B"],
    "release_year": [2016, 2017],
    "key": [20, 30]
})
df1
```

```python
df2= pd.DataFrame({
    "title": ["Movie C", "Movie D"],
    "release_year": [2018, 2019],
    "key": [40, 50]
})
df2
```

```python
pd.concat([df1, df2])
```

```python
df1 = pd.DataFrame({
    "key": [1, 2],
    "title": ["Movie A", "Movie B"]
})

df2 = pd.DataFrame({
    "key": [1, 2],
    "rating": ["PG", "R"]
})
```

```python
pd.merge(df1, df2, on="key")
```
